```python
import requests, re
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt

%matplotlib inline

#Each team has a team ID. This has to be added in the link to access the app's data.
tid = '1610612754'
game_logs = 'http://stats.nba.com/stats/teamgamelog?LeagueID=00&Season=2015-16&SeasonType=Regular+Season&TeamID=%s' % (tid)
```

```python
#Convert strings in the NBA.com datasets to integers and floats appropriately. 
def s_conv(s):
    if s[-1] == '%':
        return float(s[:-1])/100
    else:
        try:
            return float(s)
        except:
            return s
```

```python
#Calculate the spread of each game played.
def gmsprd(gid, tid):
    #Use the game ID in the game logs to find the final score for each game
    game_link = 'http://stats.nba.com/stats/boxscoresummaryv2?GameID=%s' % (gid)
    r = requests.get(game_link)
    r.raise_for_status()
    sprd = r.json()['resultSets'][5]['rowSet']
    
    s_list = [0, 0]
    
    for item in sprd:
        if int(item[3]) == int(tid):
            s_list[0] = item[-1]
        else:
            s_list[1] = item[-1]
    #Subtract the opponent's score from our team's score. A positive int means that our team won the game.
    return (s_list[0] - s_list[1])
```

```python
#Parse the game dates in the NBA.com data and convert to datetime objects. 
#Currently the date shows as 'Dec 27, 2015'
def cd(s):
    m = {'JAN': 1,
     'FEB': 2,
     'MAR': 3,
     'APR': 4,
     'MAY':5,
     'JUN': 6,
     'OCT': 10,
     'NOV': 11,
     'DEC': 12}
    d = s.replace(",","").split()
    return dt.datetime(int(d[2]), m[d[0]], int(d[1]))
```

# Finding the team IDs for all NBA teams

```python
teams_url = 'http://stats.nba.com/stats/leaguedashteamstats?Conference=&DateFrom=&DateTo=&Division=&GameScope=&GameSegment=&LastNGames=0&LeagueID=00&Location=&MeasureType=Base&Month=0&OpponentTeamID=0&Outcome=&PORound=0&PaceAdjust=N&PerMode=PerGame&Period=0&PlayerExperience=&PlayerPosition=&PlusMinus=N&Rank=N&Season=2015-16&SeasonSegment=&SeasonType=Regular+Season&ShotClockRange=&StarterBench=&TeamID=0&VsConference=&VsDivision='
tu = requests.get(teams_url)
tu.raise_for_status()
s3 = tu.json()['resultSets'][0]['rowSet']

#Create a dictionary with all teams listed by ID. 
team_dict = {}

for item in s3:
    team_dict[item[0]] = item[1]
print team_dict
current_team = team_dict[int(tid)]
print "\nCurrent team:", current_team
```

# Now to access the team stats API on NBA.com

```python
r = requests.get(game_logs)
r.raise_for_status()
gm_logs = r.json()['resultSets'][0]
#Identify the table header.
header = gm_logs['headers'][:]
print header
```

```python
#Now access all the rows in the table and put them into a DataFrame.
table = gm_logs['rowSet']
cleantable = []
for row in table:
    cleantable.append(row[:])

gamelogs_df = pd.DataFrame(cleantable, columns=header)
#Add a column to track the point spread of each game played.
gamelogs_df['SPREAD'] = gamelogs_df['Game_ID'].apply(lambda x: gmsprd(x, tid))
#Add a column and convert the dates (currently strings) to datetime objects.
gamelogs_df['CONV_DATE'] = gamelogs_df['GAME_DATE'].apply(lambda x: cd(x))
```

```python
gamelogs_df['WINS'] = gamelogs_df['WL'].apply(lambda x: 1 if x == 'W' else 0)
```

```python
titulo = '%s\nExpoloring PPG and the Spread' % (current_team)
plt.figure()
plt.subplot(211)
plt.title(titulo)
plt.plot(gamelogs_df['CONV_DATE'], gamelogs_df['SPREAD'], label='Observations', c='k', alpha=0.3)
plt.plot(gamelogs_df['CONV_DATE'], pd.rolling_mean(gamelogs_df['SPREAD'], 4), label='Rolling Mean', c='r')
plt.axhline(0, color='blue',alpha=.1, marker='.')
plt.ylabel('Point Spread')
plt.legend(loc=0, bbox_to_anchor=(1.5,1))

plt.subplot(212)
plt.plot(gamelogs_df['CONV_DATE'], gamelogs_df['PTS'], label='Observations', c='k', alpha=0.3)
plt.plot(gamelogs_df['CONV_DATE'], pd.rolling_mean(gamelogs_df['PTS'], 4), label='Rolling Mean', c='r')
plt.ylabel('Points per Game')
plt.xlabel('Game Dates')
plt.legend(loc=0, bbox_to_anchor=(1.5,1))
plt.show()
```

```python
game_df = gamelogs_df
wl = game_df[['WL', 'TOV', 'FG_PCT', 'FGA', 'STL', 'REB']].groupby('WL')
wl.agg([np.mean, np.std]) 
```

```python
plt.figure()

plt.subplot(411)
plt.title(current_team)
data = [game_df[['PTS']][game_df['WL']=='W'], game_df[['PTS']][game_df['WL']=='L']]
plt.boxplot(data, vert=False, labels=['Wins', 'Losses'])
plt.xlabel('Points per Game')
plt.show()

plt.subplot(412)
data2 = [game_df[['REB']][game_df['WL']=='W'], game_df[['REB']][game_df['WL']=='L']]
plt.boxplot(data2, vert=False, labels=['Wins', 'Losses'])
plt.xlabel('Rebounds per Game')
plt.show()

plt.subplot(413)
data2 = [game_df[['STL']][game_df['WL']=='W'], game_df[['STL']][game_df['WL']=='L']]
plt.boxplot(data2, vert=False, labels=['Wins', 'Losses'])
plt.xlabel('Steals per Game')
plt.show()

plt.subplot(414)
data3 = [game_df[['SPREAD']][game_df['WL']=='W'], game_df[['SPREAD']][game_df['WL']=='L']]
plt.boxplot(data3, vert=False, labels=['Wins', 'Losses'])
plt.xlabel('Pt Spread per Game')
plt.show()
```

# OLS on the team data

```python
import statsmodels.formula.api as smf 

df = game_df[['SPREAD', 'STL', 'REB', 'TOV']]

sp_result = smf.ols('SPREAD ~ STL + REB + TOV', data=df).fit()
print "%s\nPOINT SPREAD" % (current_team)
print sp_result.summary()
print "\n\n"

df = df = game_df[['PTS', 'FG_PCT', 'REB']]

pts_result = smf.ols('PTS ~ FG_PCT + REB', data=df).fit()
print "POINTS per GAME"
print pts_result.summary()
```

```python
titulo = "%s\nPoint Spread" % (current_team)
sp_p = sp_result.predict()
pts_p = pts_result.predict()

plt.figure()
plt.subplot(211)
plt.title(titulo)
plt.plot(game_df['CONV_DATE'], sp_p, label='predicted values')
plt.plot(game_df['CONV_DATE'], game_df['SPREAD'], c='g', alpha=.5, label='observed values')
plt.axhline(0, color='black', linestyle='-', c='k', alpha=0.3)
plt.legend(loc=1, bbox_to_anchor=(1.5,1))

plt.subplot(212)
plt.title('Points per Game')
plt.plot(game_df['CONV_DATE'], pts_p, label='predicted values')
plt.plot(game_df['CONV_DATE'], game_df['PTS'], c='g', alpha=.5, label='observed values')
plt.legend(loc=1, bbox_to_anchor=(1.5,1))

plt.show()
```
